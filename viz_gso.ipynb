```python
import matplotlib.pyplot as plt
import os
from PIL import Image

path = "/home/yoonwoo/Workspace/proj_4dgs_video/thirdparty/kubric/output/0"

fig, axs = plt.subplots(4, 4, figsize=(10, 10))

images = [] 
for i in range(16):
    images.append(Image.open(os.path.join(path, f"rgba_{str(i).zfill(5)}.png")))

for i in range(4):
    for j in range(4):
        axs[i, j].imshow(images[i*4 + j], cmap='gray')
        axs[i, j].axis('off')

plt.tight_layout()
plt.show()
```
